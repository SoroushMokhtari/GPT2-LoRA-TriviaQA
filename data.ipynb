```python
from datasets import load_dataset
from tqdm import tqdm
import pandas as pd
```

```python
dataset = load_dataset("trivia_qa", "rc")

# keyword list for filtering
keywords = [
    'movie', 'film', 'cinema', 'actor', 'actress', 'director', 'screenplay',
    'hollywood', 'oscar', 'academy award', 'box office', 'soundtrack',
    'pixar', 'disney', 'warner bros', 'universal pictures', 'paramount',
    'star wars', 'marvel', 'dc comics', 'james bond'
]

print("Filtering dataset for movie-related trivia...")
filtered_data = []

for example in tqdm(dataset['train']):
    question = example['question']
    answer = example['answer']['value']
    
    context = ""
    if example.get('entity_pages') and example['entity_pages'].get('wiki_context'):
        context = " ".join(example['entity_pages']['wiki_context'])

    if context and any(keyword in question.lower() or keyword in context.lower() for keyword in keywords):
        filtered_data.append({
            'context': context,
            'question': question,
            'answer': answer
        })

print(f"\nFound {len(filtered_data)} movie-related examples.")
```

```python
df = pd.DataFrame(filtered_data)
df.to_csv("movie_trivia_qa.csv", index=False)
```
